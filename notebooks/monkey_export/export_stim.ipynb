```python
def clean_up_structured_array(data):

    # if it is not an array, don't modify
    if not isinstance(data, np.ndarray)
        return data

   
    if data.dtype.names is not None:
        # convert record array into an array of dictionary
        value =[{k: clean_up_structured_array(entry[k]) for k in entry.dtype.names} for entry in data]
        if len(value) == 1:
            value = value[0]
    elif data.dtype.
    return value
    
```
